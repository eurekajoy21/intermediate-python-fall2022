# Scrape Sheetal Page

```
import requests
import json

import datetime

import pandas as pd
import numpy as np

from tqdm import tqdm
```

### Get Pages

```
data = {'Shape': [''],
        'clearityFrom': [''],
        'clearityTo': [''],
        'CutFrom': [''],
        'CutTo': [''],
        'PolishFrom': [''],
        'PolishTo': [''],
        'SymmetryFrom': [''],
        'SymmetryTo': [''],
        'DiaLab': [''],
        'DiaFlorcences': [''],
        'ColorFrom': [''],
        'ColorTo': [''],
        'FromDiaWt': ['0'],
        'ToDiaWt': ['20'],
        'FromPrice': ['0'],
        'ToPrice': ['999999'],
        'FromDepth': ['0'],
        'ToDepth': ['100'],
        'FromTable': ['0'],
        'ToTable': ['100'],
        'PageNo': ['0'],
        'PageSize': ['50'],
        'FancyColor': [''],
        'Rate': ['']}
```

```
response = requests.post("https://www.bluenile.com/api/public/diamond-search-grid/v2?startIndex=60&pageSize=50&_=1665377550056&unlimitedPaging=false&sortDirection=asc&sortColumn=default&shape=RD,PR,EC,AS,CU,MQ,RA,OV,PS,HS&maxDateType=MANUFACTURING_REQUIRED&isQuickShip=false&bestValueSort=false&hasVisualization=false&isFiltersExpanded=false&astorFilterActive=false&country=USA&language=en-us&currency=USD&productSet=BN")
```

```
response
```

```
response = requests.post("https://www.sheetaldiamonds.com/api/RestApi/GetDiamondData", data=data)
```

```
total_diamonds = int(json.loads(response.json())['ResponseData']['Count'][0]['Count'])
total_diamonds
```

```
pages = np.arange(0, total_diamonds, 50)
```

```
pages
```

```
diamonds = []
for page in tqdm(pages[:5]):
    data['PageNo'] = [f'{page}']
    response = requests.post("https://www.sheetaldiamonds.com/api/RestApi/GetDiamondData", data=data)
    response_diamonds = json.loads(response.json())['ResponseData']['dtDiamond']

    diamonds.extend(response_diamonds)
```

```
df = pd.DataFrame(diamonds)
```

```
df
```

```
today = datetime.datetime.today().strftime("%Y-%m-%d")
```

```
df.to_csv(f"data/diamonds-{today}.csv", index=False)
```
